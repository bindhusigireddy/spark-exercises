```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
spark=SparkSession.builder.appName("spark_learning").getOrCreate()
```

```python
data = [
  ('Jas','','Smith','1991-04-01','M',3000),
  ('Michael','Rose','','2000-05-19','M',4000),
  ('Robert','','Williams','1978-09-05','M',4000),
  ('Maria','Anne','Jones','1967-12-01','F',4000),
  ('Jen','Mary','Brown','1980-02-17','F',-1)
]
```

```python
columns= ["Firstname", "middlename", "lastname", "dob", "gender", "salary"]
df=spark.createDataFrame(data, columns)
```

```python
df.show()
```

```python
df.withColumnRenamed("dob", "DateofBirth").show()
```

```python
df.withColumn("salary",col("salary").cast("Integer")).show()
```

```python
df.withColumn("salary",col("salary")* 100).show()
```

```python
df.withColumn("checking", col("salary")* -1).show()
```

```python
df.drop("checking").show()
```

```python
df.show(truncate=False)
```

```python
salary = df.filter(df.salary == 3000)
salary.show()
```

```python
salary = df.filter(df.salary!=3000)
salary.show()
```

```python
a="Rose"
df.filter(df.middlename.isin(a)).show()
```

```python
df.filter(df.middlename.startswith("A")).show()
```

```python
df.filter(df.lastname.endswith("s")).show()
```

```python
df.filter(df.Firstname.contains("R")).show()
```

```python
df.filter(df.lastname.like("%W%")).show()
```

```python
df.filter(df.Firstname.rlike("(?i)^*rose$")).show()
```

```python
simpleData = [("James","Sales","NY",90000,34,10000), 
    ("Michael","Sales","NY",86000,56,20000), 
    ("Robert","Sales","CA",81000,30,23000), 
    ("Maria","Finance","CA",90000,24,23000), 
    ("Raman","Finance","CA",99000,40,24000), 
    ("Scott","Finance","NY",83000,36,19000), 
    ("Jen","Finance","NY",79000,53,15000), 
    ("Jeff","Marketing","CA",80000,25,18000), 
    ("Kumar","Marketing","NY",91000,50,21000) 
]
```

```python
columns=["employee_name","department","state","salary","age","bonus"]
```

```python
df2=spark.createDataFrame(simpleData, columns)
```

```python
df2.show()
```

```python
df2.sort(col("department"),col("employee_name")).show()
```

```python
df2.orderBy(col("department").asc(),col("state").desc()).show()
```

```python
df2.orderBy(col("employee_name").asc(), col("bonus").desc()).show()
```

```python

df2.createOrReplaceTempView("emp")
a = spark.sql("""
  SELECT employee_name, department, state, salary, age, bonus
  FROM emp
  ORDER BY bonus DESC
""")
a.show()

```
